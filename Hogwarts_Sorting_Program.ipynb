```python
#3.0 Version (8 Questions)
ques_tuple = ('''
    1. Which of the following would you most hate people to 
       call you?
    A:Ordinary
    B:Ignorant
    C:Cowardly
    D:Selfish''',
'''
    2. Given the choice, would you rather invent a potion that
       would guarantee you:
    A:Glory
    B:Wisdom
    C:Love
    D:Power''',
'''
    3. After you have died, what would you most like people to
       do when they hear your name?
    A:Miss you, but smile
    B:Ask for more stories about your adventures
    C:Think with admiration of you achievements
    D:I don't care what people think of me after I'm dead; it's
      what they think of me while I'm alive that counts''',
'''
    4. How would you like to be known to history?",
    A:The Wise
    B:The Good
    C:The Bold
    D:The Great''',
'''
    5. You enter an enchanted garden. What would you be most 
       curious to examine first?",
    A:The silver leafed tree bearing golden apples,
    B:The fat red toadstools that appear to be talking to each other,
    C:The bubbling pool, in the depths of which something luminous
      is swirling,
    D:The statue of an old wizard with a strangely twinkling eye''',
'''
    6. Four boxes are placed before you. Which would you 
       try and open?,
    A:"The small tortoiseshell box, embellished with gold, inside 
       which some small creature seems to be squeaking.
    B:"The gleaming jet black box with a silver lock and key,
       marked with a mysterious rune that you know to be the mark of Merlin.
    C:" The ornate golden casket, standing on clawed feet, whose inscription
       warns that both secret knowledge and unbearable temptation lie within.
    D:" The small pewter box, unassuming and plain, with a scratched
       message upon it that reads I open only for the worthy.''',
'''
    7. Four goblets are placed before you. Which would you 
       choose to drink?
    A:The foaming, frothing, silvery liquid that sparkles as 
      though containing ground diamonds.
    B:The smooth, thick, richly purple drink that gives off a
      delicious smell of chocolate and plums.
    C:The golden liquid so bright that it hurts the eye, and
      which makes sunspots dance all around the room.
    D:The mysterious black liquid that gleams like ink, and gives
      off fumes that make you see strange visions.''',
'''
    8. What kind of instrument most pleases your ear?
    A:Violin
    B:Drum
    C:Piano
    D:Trumpet''')
choice_tuple = ({'A':'Hufflepuff','B':'Ravenclaw','C':'Slytherin','D':'Gryffindor'},
        {'A':'Hufflepuff','B':'Gryffindor','C':'Ravenclaw','D':'Slytherin'},
        {'A':'Ravenclaw','B':'Hufflepuff','C':'Gryffindor','D':'Slytherin'},
        {'A':'Gryffindor','B':'Hufflepuff','C':'Ravenclaw','D':'Slytherin'},
        {'A':'Slytherin','B':'Hufflepuff','C':'Hufflepuff','D':'Ravenclaw'},
        {'A':'Ravenclaw','B':'Slytherin','C':'Hufflepuff','D':'Gryffindor'},
        {'A':'Slytherin','B':'Hufflepuff','C':'Ravenclaw','D':'Gryffindor'},
        {'A':'Slytherin','B':'Hufflepuff','C':'Ravenclaw','D':'Gryffindor'})

house_tuple = ( '''
            Your test scores show high levels of Bravery and chivalry,
            which is valued in the Gryffindor House.''',
          '''
            Your test scores show high levels of ambition, cunning,
            leadership, and resourcefulness, which is valued in the 
            Slytherin House.''',
          '''
            Your test scores show high levels of intelligence, knowledge,
            and wit, which is valued in the Ravenclaw House.''',
          '''
            Your test scores show high levels of patience, loyalty, 
            and fair-play, which is valued in the Hufflepuff House.''')

score_dict = {"Gryffindor":[0,house_tuple[0]],"Slytherin":[0,house_tuple[1]],
        "Ravenclaw":[0,house_tuple[2]],"Hufflepuff":[0,house_tuple[3]]}

def ask_ques(i):
  while True:
    print("*" * 80)
    print(ques_tuple[i])
    print("*" * 80)
    answer = input("Please input your answer: (A,B,C,D)")
    if answer in ("A","B","C","D"):
      score_dict[choice_tuple[i][answer]][0] += 1
      break
    else:
      print("Input error! Please try again.")


for i in range(8):
  ask_ques(i)

key_list = score_dict.keys()
highest = "Slytherin"
for key in key_list:
  if score_dict[highest][0] < score_dict[key][0]:
    highest = key
  print("%s: %s" %(key,score_dict[key][0]))
print("=" * 70)
print("\t\t\t CONGRATULATIONS")
print("\t\tYou have been admitted to House %s" %highest)
print("\t     Of Hogwarts School of Witchcraft and Wizardry")
print(score_dict[highest][1])
print("=" * 70)
```
